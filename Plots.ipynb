```python

#%%
from numpy.lib.function_base import diff
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


sweat = pd.read_csv('Final Results/sweat_FinalResults.csv')
breast = pd.read_csv('Final Results/breast_FinalResults.csv')
urban = pd.read_csv('Final Results/urban_FinalResults.csv')
sweat_ordinal = pd.read_csv('Final Results/ordinal_FinalResults.csv')
parkinson = pd.read_csv('Final Results/parkinsons_FinalResults.csv')
community = pd.read_csv('Final Results/community_FinalResults.csv')


            
datasets = [{'name': 'Wisconsin Breast Cancer','Data':breast},
            {'name': 'Urban Land Cover','Data':urban},
            {'name': 'Sweat Study','Data':sweat},
            {'name': "Parkinson's",'Data':parkinson},
            {'name': 'Communities and Crimes','Data':community},
            {'name': 'Sweat Study (Regression)','Data':sweat_ordinal},]
 
plt.rcParams["font.family"]= "serif",
plt.rcParams['axes.edgecolor'] = "0.15"
plt.rcParams['axes.linewidth'] = 1.00
plt.rcParams['font.size'] = 20
plt.rcParams['text.usetex'] = True
plt.rcParams['text.latex.preamble']=[
        r"""
        \usepackage{libertine}
        \usepackage[libertine]{newtxmath}
        \usepackage{amsmath}
        """]
sns.set_style("whitegrid", {'axes.grid' : False})

fig, axes = plt.subplots(2,3,figsize=(19, 8),squeeze=True,frameon = True)
for data,ax in zip(datasets,axes.flat):
    name,df = data['name'],data['Data']
    ax = sns.lineplot(df['k'],df['PCD'],color="black",marker='o',data=df,ci=None,ax=ax)
    ax.set(xlabel=r'$\mathit{k}$', ylabel=r'PCD',title=r"{%s}" % name)
    ax.autoscale()
    ax.legend([],[], frameon=False)
    ax.spines['bottom'].set_color('0.5')
    ax.spines['top'].set_color('0.5')
    ax.spines['right'].set_color('0.5')
    ax.spines['left'].set_color('0.5')
plt.tight_layout()
fig.savefig('Images/sensitivity_analysis.png',dpi=500)
```
